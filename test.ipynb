```python
from torch.nn.modules.batchnorm import BatchNorm2d

def bn_freeze(model):
    if type(model) is BatchNorm2d:
        for param in model.parameters():
            param.requires_grad = False
    for child in model.children():
        bn_freeze(child)
```

```python
bn_freeze(model)
```
